#RNN have a forget problem

In the end a simple RNN is a chain of functions (or a multiplication in backpropagation). There are two problems associated.
* During backpropagation, if one of the factors it's 0 or close to 0 the gradients vanish. They can also explode.
* For inference, the RNN tends to forget things that are far apart in a sentence/document.

$z(t), r(t), h(t)$ are vectors of size M. M is hyperparameter.
Circle with dot is element-wise multiplication (like a dot product but without adding at the end).
$z(t)$ tells me if I want to remember the previous term ($(1-z_t) → 1$) or keep the new one ($z_t → 1$).

# They are basically a binary classifier. They choose to forget or remember a word.
